```python
def read_file(file_name,folder_contain,validation=True):
    if validation :
        with open(f'{folder_contain}/gt_{file_name}.txt',"r") as f:
            data = f.read().replace('','')
    else:
        with open(f'{folder_contain}/{file_name}.txt',"r") as f:
            data = f.read().replace('','')
    data = data.split('\n')
    data.remove('')
    data.sort()
    return data
    
```

```python
def handle_data(list_points):
    polygon_list = []
    for index in list_points:
        temp_point_split = index.split(',')
        temp_point = [float(temp_point_split[x]) for x in range(8)]
        polygon = Polygon([(temp_point[0],temp_point[1]),(temp_point[2],temp_point[3]), \
                                      (temp_point[4],temp_point[5]),(temp_point[6],temp_point[7])])
        if len(temp_point_split) == 9:
            polygon_list.append([polygon,temp_point_split[8]])
        else :
            polygon_list.append([polygon,' '])
    return polygon_list
```

```python
def IOU(polygon_list_1, polygon_list_2):
    # polygon_list_1 : list from validation data
    # polygon_list_2 : list from predicted data
    result = []
    for index in polygon_list_1:
        flag = False
        if index[1] == '###':
            pass
        else:
            for index_2 in polygon_list_2:
                if index[0].intersects(index_2[0]):
                    flag = True
                    intersect = index[0].intersection(index_2[0]).area
                    union = index[0].union(index_2[0]).area
                    iou = intersect / union
                    x,y = index[0].exterior.xy
                    x1,y1 = index_2[0].exterior.xy
                    #plt.plot(x,y)
                    #plt.plot(x1,y1)
                    if len(index[1]) !=0:
                        precision = len([(i, j) for i,j in zip(index[1],index_2[1]) if i==j]) / len(index_2[1])
                        recall = len([i for i in index_2[1] if i in index[1]]) / len(index_2[1])
                        if recall > 1 or precision >1:
                            print(f'precision: {precision} -  recall: {recall} -  Text :{index[1]} - predicted: {index_2[1]} - IOU :{iou}')
                    else:
                        precision = 0
                        recall = 0
                    print(f'precision: {precision} -  recall: {recall} -  Text :{index[1]} - predicted: {index_2[1]} - IOU :{iou}')
                    result.append([index[0],index_2[0],iou,index[1],index_2[1],precision,recall])
            if not flag:
                result.append([index[0],None,0,index[1],None,0,0])
    return result
```

```python
data_ICDAR = read_file("img_1","validation",True)
data_lanyOCR =  read_file("img_1","outputs",False)
poly = handle_data(data_ICDAR)
poly_1 = handle_data(data_lanyOCR)
```

```python
def output_file(file_name,folder_contain, list_result):
    with open(f'{folder_contain}/{file_name}_result.txt','w') as f:
        for result in list_result:
            x,y = result[0].exterior.xy
            for index in range(len(x)):
                f.write(str(x[index])+',')
                f.write(str(y[index])+',')
            f.write(result[3])
            f.write(' - ')
            if result[1] is not None:
                x1,y1  = result[1].exterior.xy
                for index in range(len(x1)):
                    f.write(str(x1[index])+',')
                    f.write(str(y1[index])+',')
                f.write(result[4])
               
            else:
                f.write('-1')
            f.write(f' - IOU : {result[2]} - precision - {result[5]} - recall - {result[6]}\n')
```

```python
result = IOU(poly,poly_1)
output_file('test.txt','IOU',result)
```

```python
result_IOU = []
result_precision = []
result_recall = []
for index in name:
    data_ICDAR = read_file(index,"validation",True)
    data_lanyOCR =  read_file(index,"outputs",False)
    point_ICDAR = handle_data(data_ICDAR)
    point_lanyOCR = handle_data(data_lanyOCR)
    result = IOU(point_ICDAR,point_lanyOCR)
    for value in result:
        result_IOU.append(value[2])
        result_precision.append(value[5])
        result_recall.append(value[6])
    output_file(index,'IOU',result)
```

```python
final_IOU = array(result_IOU)
```

```python
sns.distplot(final_IOU,bins=30)
print(final_IOU.mean())
```
